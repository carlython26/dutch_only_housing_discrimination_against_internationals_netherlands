```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
base_url = "https://kamernet.nl"
all_rooms = []

for link in links_df['links']:
    url = base_url+link
    print(url)
    rooms_html = requests.get(url).text
    rooms_soup = BeautifulSoup(rooms_html)
    rooms_language = rooms_soup.select(".mui-style-1fsfdy1")
    
    try:
        # print(rooms_language[9].text.strip()) 
        language = rooms_language[9].text.strip()
        print('-------')
    except Exception as e:
        print(f"An error occured: {e}")
        language = 0
        print('-------')

    temp_room = {
        'url': url,
        'language': language
    }

    all_rooms.append(temp_room)

    time.sleep(1)

all_rooms

# how to structure this data into a list of dictionaries
# because dataframe is list of dictionaries
```
